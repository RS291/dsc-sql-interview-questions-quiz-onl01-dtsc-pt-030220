## Question 4

Explain the various types of JOINs possible with SQL. 

Type your answer below this line:
_______________________________________________________________________________________________________________________________
There are left and right joins which use the respective entire table to join with another table but only using shared data from the latter table. An inner join only returns what is shared between both tables. An outer join returns only data that is not present in both tables.
